```python
final_df = df[['Escalation Phrases','Reopen','Response Time in min','Reopen Counter','Total Comments','Total Agents','Case Age','Instance','MilestoneIcon','ActiveServiceAgreement','CaseReason','AtRisk','ATEscalation']]
# final_df = df[['Reopen','Reopen Counter','Total Comments','Total Agents','AccountName','CaseReason','Feature','ActiveServiceAgreement','AtRisk','SubReason','Priority','Instance','MilestoneIcon','SubFeature','Theater','ATEscalation']]
#CaseOrigin
```

```python
cols = ['Reopen','Instance','ActiveServiceAgreement','CaseReason','AtRisk','MilestoneIcon']
# cols = ['Reopen','AccountName','CaseReason','Feature','ActiveServiceAgreement','AtRisk','SubReason','Priority','Instance','MilestoneIcon','SubFeature','Theater']
for c in cols:
    final_df[c].fillna('Unknown',inplace=True)
```

```python
# final_df['smx_case_no_of_days'].fillna(method='ffill',inplace=True)
```

```python
final_df['Response Time in min'] = final_df.groupby("ATEscalation")['Response Time in min'].transform(lambda x: x.fillna(x.mean()))
final_df['Case Age'] = final_df.groupby("ATEscalation")['Case Age'].transform(lambda x: x.fillna(x.mean()))
```

```python
mean = final_df['Response Time in min'].mean()
std = final_df['Response Time in min'].std()
final_df['Response Time in min'] = final_df['Response Time in min'].apply(lambda x:(x-mean)/std)
```

```python
mean = final_df['Case Age'].mean()
std = final_df['Case Age'].std()
final_df['Case Age'] = final_df['Case Age'].apply(lambda x:(x-mean)/std)
```

```python
cols = ['Reopen','CaseReason','Instance','ActiveServiceAgreement','AtRisk','MilestoneIcon']
# cols = ['Reopen','AccountName','CaseReason','Feature','ActiveServiceAgreement','AtRisk','SubReason','Priority','Instance','MilestoneIcon','SubFeature','Theater']
temp_df = final_df[['Reopen','Instance','ActiveServiceAgreement','AtRisk','MilestoneIcon']]
# temp_df = final_df[['Reopen','AccountName','CaseReason','Feature','ActiveServiceAgreement','AtRisk','SubReason','Priority','Instance','MilestoneIcon','SubFeature','Theater']]
enc = OneHotEncoder(handle_unknown='ignore')
enc.fit(temp_df.values)
temp_df = pd.DataFrame(enc.transform(temp_df.values).toarray())
for c in cols:
    final_df.drop(columns=[c],inplace=True)
final_df = final_df.join(temp_df)
final_df
```

```python
y = final_df['ATEscalation']
final_df.drop(columns=['ATEscalation'],inplace=True)
```

```python
y_s = y.ravel()
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train, X_test, y_train, y_test = train_test_split(final_df.values, y, test_size=0.2,stratify = y_s)
```

```python
from sklearn.linear_model import LogisticRegression
log_clf = LogisticRegression(random_state=0, solver='liblinear',max_iter=200,C=0.5)
log_fit = log_clf.fit(X_train,y_train)
pred = log_clf.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(max_depth=25,min_samples_split=4,criterion = 'entropy',random_state=0)
rf_clf.fit(X_train,y_train)
pred = rf_clf.predict(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
svm_clf = make_pipeline(StandardScaler(), SVC(gamma='auto',probability=True))
svm_clf.fit(X_train,y_train)
```

```python
import xgboost as xgb
xg_clf = xgb.XGBClassifier(booster='dart',objective="binary:logistic", random_state=42)
xg_clf.fit(X_train,y_train)
pred = xg_clf.predict(X_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import make_classification
# X, y = make_classification(n_samples=1000, n_features=4,n_informative=2, n_redundant=0,random_state=0)
ada_clf = AdaBoostClassifier(n_estimators=100, random_state=0)
ada_clf.fit(X_train,y_train)
pred = ada_clf.predict(X_test)
```
